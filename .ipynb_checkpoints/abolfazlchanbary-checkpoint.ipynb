```python

varin=[]
unoque=[]
count=[]
dtype=[]
missing=[]

for x in df.columns:
    varin.append(x)
    unoque.append(len(df[x].unique()))
    count.append(len(df[x]))
    dtype.append(df[x].dtype)
    missing.append(df[x].isna().sum())
    missing_percent = [round((m / c) * 100, 2) if c > 0 else 0 for m, c in zip(missing, count)]
    


output=pd.DataFrame({
"Variable":varin,
"unique":unoque,
"count":count,
"dtype":dtype,
"missing":missing,
"missing_percent":missing_percent
})
display(output)
```

```python

display(HTML('<span style="color:red;font-size:25px">Click a column to view unique values</span>'))


df = pd.read_csv("heart.csv")

# Create an output widget to display results
output = widgets.Output()

def show_uniques(column):
    with output:
        # Clear the previous output before displaying the new one
        output.clear_output()

        # Show the unique values of the selected column
        uniques = df[column].unique().tolist()
        
        # Format the unique values with bullet points and line breaks
        uniques_str = 'ðŸ”¹ '.join([f'<span style="color: black;">{str(item)}</span>' for item in uniques])
        html_text = f"""
    <h3 style="text-align:center;">{'-'*10} <span style="color:black;text-shadow:0px 0px 12px #04b54a">{column}</span> {'-'*10}</h3>
    <span style="color:blue; font-weight:bold;">ðŸ’™ Unique values:</span><br>ðŸ”¹ {uniques_str}
    """
        display(HTML(html_text))

# Create a button for each column
buttons = []
for column in df.columns:
    button = widgets.Button(description=column)
    button.on_click(lambda b, col=column: show_uniques(col))  # Pass column name to the callback
    buttons.append(button)

# Layout to display the buttons horizontally
button_layout = HBox(buttons)

# Create a vertical layout for buttons and output
layout = VBox([button_layout, output])

# Display the layout with buttons and output
display(layout)

```
